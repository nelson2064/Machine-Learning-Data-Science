## Matplotlib example workflow

```python

# A matplotlib workflow

# 0. Import and get matplotlib ready
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup plot

fig, ax = plt.subplots(figsize=(10,10))


# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("../images/simple-plot.png")
```

### Histogram (hist)

* Could show image of normal distribution here

```python

# Generate some continuous data
data = np.random.normal(100, 15, 1000)

# Plot the histogram
fig, ax = plt.subplots()
ax.hist(data, bins=50, color='blue', alpha=0.7)
ax.set(title='Histogram of Normally Distributed Data', xlabel='Data Values', ylabel='Frequency')
plt.show()
```

### which one should you use? (pyplot vs matplotlib OO method?)

when plotting somethign, queickly its okay to use the pyplot method

when plotting something more advanced, use the oo method

For more complicated plots, you'll want to use the OO method.

```python
heart_disease.head()
```

```python
# Perform data analysis on patients over 50
over_50 = heart_disease[heart_disease["age"] > 50]
len(over_50)
```

```python
over_50.head()
```

```python
#pyplot method

#does this looks okay its doing what we want do but its not great so let push it more so how we can do that we can do thorugh oo method


# color = target
over_50.plot(kind="scatter",
            x='age',
             y='chol',
             c='target'
            );
```

## OO method

```python
#oo method mix with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter', 
             x="age", 
             y="chol", 
             c='target', 
             ax=ax);
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter', 
             x="age", 
             y="chol", 
             c='target', 
             ax=ax);
ax.set_xlim([45, 100]);
```

```python
over_50.target.values
```

```python
## OO method from scratch
# Make a bit more of a complicated plot

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");
            #go to the scatter variable value and then find the .legend_elements() whcih it gets from c parameter
ax.legend(*scatter.legend_elements(), title="Target");

```

## so now this is looking bit better oo method from scratch looking far much better then up we built previously this is how few line of code you now seee how customize this oo api is what if we want to add more few thing so we can do that what if we want something like horizonal line to indicate the average colestrol

```python
## OO method from scratch
# Make a bit more of a complicated plot

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(over_50["age"], 
                     over_50["chol"], 
                     c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");
            #go to the scatter variable value and then find the .legend_elements() whcih it gets from c parameter
ax.legend(*scatter.legend_elements(), title="Target");

## Add a horizonatal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--'
          )
```
